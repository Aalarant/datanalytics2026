```python
import numpy as np
import pandas as pd

# let's see what the data says
df = pd.read_csv('loans.csv')
df.head()

```

```python
# Let's filter out exceedingly comical loans
df = df[df["Current Loan Amount"] < 9999999]
```

```python
# Let's see what loan(s) have the highest credit balance 
df.nlargest(1, "Current Credit Balance")
```

```python
# Let's see which loan has the lowest annual income
df.nsmallest(1, "Annual Income")
```

```python
# Let's see the min, max and mean of current loans.
pd.set_option('display.float_format', '{:.2f}'.format)
df["Current Loan Amount"].describe()

# Min is 15422.00
# Max is 789250.00
# And mean is 312122.30
```

```python
# Let's create a new field:
df["Actual Annual Income"] = df["Annual Income"] - 12 * df["Monthly Debt"]
```

```python
# How many loans are "Long Term"?
df["Term"].value_counts()

# Answer: 16752
```

```python
# What are the 3 most common loan purposes?
df["Purpose"].value_counts().head(3)

# Debt Consolidation, Other and Home Improvements
```

```python
# How many unique home ownerships are there?
df["Home Ownership"].unique()
```
