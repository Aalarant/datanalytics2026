```python
# Let's load the dataset:
import pandas as pd

df = pd.read_csv("furniture.csv")
df.head()
```

```python
# Let's inspect duplicate rows:
duplicate_count = df.duplicated().sum()
duplicate_count

# The amount of duplicate rows is 110
```

```python
# Let's remove the duplicate rows:
df = df.drop_duplicates()
```

```python
# Now we inspect the NaN values:
# Since I didn't know how to do this I asked CoPilot for help. 
# This is the following result:
nan_count = df.isna().sum()
nan_count
```

```python
# Now we remove all rows that contain these values
df_clean = df.dropna()
```

```python
# Let's create a new column:
def categorize_delivery(days):
    if days < 4:
        return "quick"
    elif 4 <= days <= 7:
        return "normal"
    else:
        return "slow"

#Let's apply this to the dataset:
df["delivery_category"] = df["delivery_days"].apply(categorize_delivery)

# Let's inspect:
df["delivery_category"].value_counts()
```

```python
# Lets preview now the dataset:
df.head()
```

```python
# After previewing, we shall remove the original delivery_days columN:
df = df.drop(columns=["delivery_days"])
```
